```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
def analyze_sentiment(data):
    analysis = SIA().polarity_scores(data)
    if analysis['compound'] > 0:    
        return 'Positive'
    elif analysis['compound'] == 0: 
        return 'Neutral'
    else:                           
        return 'Negative'
```

```python
df['sentiment_neutral_ratings'] = df[df.score == 3].content.apply(analyze_sentiment)
```

```python
df=df[df.sentiment_neutral_ratings != 'Neutral']
```

```python
def rating_seg(data):
    score_ = 0
    if df['score'] <= 2:
        score_ = 0
    elif df['score'] == 3:
        if data['sentiment_neutral_ratings'] == 'Positive':
            score_ = 0
        else:
            score_ = 1
    else:
        score_ = 1
    return score_
```

```python
def rating_seg(data):
    score_ = 0
    score_bool = data['score'].bool()
    if score_bool:
        if data['score'] <= 2:
            score_ = 0
        elif data['score'] == 3:
            score_ = 1
        elif data['score'] == 4:
            score_ = 2
        else:
            score_ = 3
    return score_
```

```python
def rating_seg(data):
    score_ = 0
    score_int = data['score']
    if score_int:
        if score_int <= 2:
            score_ = 0
        elif score_int == 3:
            score_ = 1
        elif score_int == 4:
            score_ = 2
    else:
        score_ = 3
    return score_
```

```python
df['new_score'] =df.apply(rating_seg, axis=1)
df[['score','new_score']].sample(frac=1.0).head() 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Tfidf_vect = TfidfVectorizer()
Train_content_tfidf = Tfidf_vect.fit_transform(df['new_content'].astype(str))
```

```python
print(Train_content_tfidf[:3])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.model_selection import StratifiedKFold,cross_val_score
```

```python
SKF = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
for model in [LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier(),SVC()]:
    model_name = str(model).split('(',1)[0]
    print(f' Accuracy of {model_name}  : {round(cross_val_score(model,Train_content_tfidf,df.new_score,cv=SKF,scoring="accuracy").mean()*100,2)}%')
```
